```python
print("========== STRING OPERATIONS ==========")
```

```python
# Taking input from user
text = input("Enter a string: ")
```

```python
# 1. Length of String
print("\n1. Length of String:", len(text))
```

```python
# 2. Uppercase
print("2. Uppercase:", text.upper())
```

```python
# 3. Lowercase
print("3. Lowercase:", text.lower())
```

```python
# 4. Capitalize
print("4. Capitalize:", text.capitalize())
```

```python
# 5. Title Case
print("5. Title Case:", text.title())
```

```python
# 6. Swap Case
print("6. Swap Case:", text.swapcase())
```

```python
# 7. Reverse String
print("7. Reverse String:", text[::-1])
```

```python
# 8. Count a Character
ch = input("\nEnter a character to count: ")
print("Count:", text.count(ch))
```

```python
# 9. Find a Substring
sub = input("\nEnter substring to find: ")
```

```python
old = input("\nEnter word to replace: ")
new = input("Enter new word: ")
print("After Replace:", text.replace(old, new))
```

```python
# 11. Startswith
start = input("\nEnter starting word: ")
print("Starts with:", text.startswith(start))
```

```python
# 12. Endswith
end = input("Enter ending word: ")
print("Ends with:", text.endswith(end))
```

```python
# 13. Split String
print("\nSplit List:", text.split())
```

```python
# 14. Join Example
joined = "-".join(text.split())
print("Joined with '-':", joined)
```

```python
# 15. Remove Extra Spaces
print("Without Extra Spaces:", " ".join(text.split()))
```

```python
# 16. Check String Properties
print("\n----- String Property Checks -----")
print("Is Alphabetic:", text.isalpha())
print("Is Numeric:", text.isdigit())
print("Is Alphanumeric:", text.isalnum())
print("Is Lowercase:", text.islower())
print("Is Uppercase:", text.isupper())
print("Is Space:", text.isspace())
print("Is Title:", text.istitle())
```

```python
# 17. Indexing
print("\n----- Indexing -----")
if len(text) > 0:
    print("First Character:", text[0])
    print("Last Character:", text[-1])
```

```python
# 18. Slicing
print("\n----- Slicing -----")
print("First 3 Characters:", text[:3])
print("Last 3 Characters:", text[-3:])
print("Every Second Character:", text[::2])

print("\n========== END OF PROGRAM ==========")
```
