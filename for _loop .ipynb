# for loop


```python
# =============python for loop practice
```

```python
#print number from 1 to 10
for i in range(1,11):
    print (i)
```

```python
#print even number from 1 to 20
for i in range(2,21,2):
    print(i)
```

```python
#print odd number from 1 to 20
for i in range(1,20,2):
    print(i)
```

```python
#multiplication table for 5
print("\n multiplication table of 5:")
for i in range(1,11):
    print(f" 5 x{i}={5*i}")
```

```python
#sum of numbers from 1 -10
print("sum of 1 to 10\n")
total=0
for i in range(1,11):
    total+=i
    print(total)
```

```python
#count from 10 to 1
for i in range(10,0,-1):
    print(i)
```

```python
#print the letter of a word
word="PYTHON"
for letter in word:
    print(letter)
```

```python
#print ASCII values of A to E
for ch in "ABCDE":
    print(ch,"=",ord(ch))
```

```python
#count vovels in a string
text="education"
count=0
for ch in text:
    if ch in 'aeiou':
        count+=1
print("vowels are:",count)
```

```python
#print number divisible by 5 from 1 to 50
for i in range(1,101):
    if i%5==0:
        print(i)
```

```python
#print cubes from 1 to 25
for i in range(1,26):
    print(i**3)
```
