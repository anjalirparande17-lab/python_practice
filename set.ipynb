```python
#set operations
#intersection
A={1,2,3,4,5}
B={4,5,6,7,8}
A.intersection(B)
```

```python
#union
A.union(B)
```

```python
#difference
A.difference(B)
```

```python
B.difference(A)
```

```python
A.union(B)-A.intersection(B)
```

```python
A={1,2,3,4,5}
B={4,5,6,7,8}
res=A.symmetric_difference(B)
print(res)
```
